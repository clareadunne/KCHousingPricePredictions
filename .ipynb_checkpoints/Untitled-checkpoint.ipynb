```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import scipy.stats as st
warnings.filterwarnings("ignore") 
pd.set_option('display.max_columns', 360)
```

```python
df = pd.read_csv('TotalSearchResults.csv')
```

```python
homes = df[df['Living Area'] != 0]
```

```python
homes['livarea'] = homes['Living Area']

homes
```

```python
from statsmodels.formula.api import ols

ols(formula='Price~livarea', data=homes).fit().summary()
```
